```python
from obspy.iris import Client
from obspy import UTCDateTime
client = Client()
```

```python
t = UTCDateTime("2010-02-27T06:45:00.000")
st = client.getWaveform("IU", "ANMO", "00", "BHZ", t, t + 60 * 60)
st.plot()
```

```python
event = client.getEvents(eventid=4099810)[0]
```

```python
event_time = event.preferred_origin().time
time_window = [event_time - 60*5, event_time + 60*60]
```

```python
availability = client.availability(channel='BH?,LH?', starttime=time_window[0], endtime=time_window[1])
available_nslc = [line.split()[:4] for line in availability.split('\n')]
print len(available_nslc)
print available_nslc[0]
```
